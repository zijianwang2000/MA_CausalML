```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

## RMSE of ML models

```python
with open('results_dict_1.pkl', 'rb') as pickle_file:
    results_dict_ = pickle.load(pickle_file)

results_dict_1 = {}

for N, results in results_dict_.items():
    rmses = results[-1]
    results_dict_1[N] = np.mean(rmses[:, 3:], axis=0)

with open('results_dict_2.pkl', 'rb') as pickle_file:
    results_dict_ = pickle.load(pickle_file)

results_dict_2 = {}

for N, results in results_dict_.items():
    rmses = results[-1]
    results_dict_2[N] = np.mean(rmses[:, 3:], axis=0)
```

```python
for N in results_dict_1.keys():
    print(results_dict_1[N], results_dict_2[N])
```

```python
sample_sizes = list(results_dict_1.keys())
rmse = np.array(list(results_dict_1.values()))

for k in range(3):
    plt.plot(np.log(sample_sizes), np.log(rmse[:,k]))
```

```python
slopes = np.polyfit(np.log(sample_sizes), np.log(rmse), 1)[0,:]
-slopes
```

## RMSE of DML estimator

```python
theta_0 = 1.0
```

```python
with open('results_dict_1.pkl', 'rb') as pickle_file:
    results_dict_ = pickle.load(pickle_file)

results_dict_1 = {}

for N, results in results_dict_.items():
    ate_estimates = results[0]
    results_dict_1[N] = np.sqrt(np.mean((ate_estimates[:, 5]-theta_0)**2, axis=0))

with open('results_dict_2.pkl', 'rb') as pickle_file:
    results_dict_ = pickle.load(pickle_file)

results_dict_2 = {}

for N, results in results_dict_.items():
    ate_estimates = results[0]
    results_dict_2[N] = np.sqrt(np.mean((ate_estimates[:, 5]-theta_0)**2, axis=0))
```

```python
for N in results_dict_1.keys():
    print(results_dict_1[N], results_dict_2[N])
```
