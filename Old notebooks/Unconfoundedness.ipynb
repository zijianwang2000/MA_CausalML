```python
import numpy as np
import matplotlib.pyplot as plt
from data_generation import get_data
```

```python
list_ = []

for j in range(5000):
    y, d, x = get_data(8000)
    tau_dm = np.sum(y*d)/np.sum(d) + np.sum(y*(1-d))/np.sum(1-d)
    list_.append(tau_dm)
```

```python
plt.hist(list_)
plt.show()
```

This shows that the difference-in-means estimator $\hat{\tau}_{\text{dm}}$ is not consistent for the ATE (value 1). So treatment assignment is indeed non-randomized.
